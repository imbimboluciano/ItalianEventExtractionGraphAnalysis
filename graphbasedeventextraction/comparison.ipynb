```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from util import perfomance_evaluator 
```

## Comparison on Italian datasets

```python
evaluator = perfomance_evaluator.PerformanceEvaluator("it","../dataset/cleaned_tweets/cleaned_italian_tweets.csv")
all_events = evaluator.get_all_events("../dataset/event_extracted/stanza/event_extracted_italian_stanza.csv")
accuracy_stanza = evaluator.accuracy(all_events)
print(f'Accuracy: {accuracy_stanza*100}%')
```

```python
number_of_events_groundtruth = evaluator.get_n_events_groundtruth()
number_of_events_extracted_stanza = evaluator.get_n_events_classified()
```

```python
evaluator = perfomance_evaluator.PerformanceEvaluator("it","../dataset/cleaned_tweets/cleaned_italian_tweets.csv")
all_events = evaluator.get_all_events("../dataset/event_extracted/spacy/event_extracted_italian_spacy.csv")
accuracy_spacy = evaluator.accuracy(all_events)
print(f'Accuracy: {accuracy_spacy*100}%')
```

```python
number_of_events_extracted_spacy = evaluator.get_n_events_classified()
```

```python
evaluator = perfomance_evaluator.PerformanceEvaluator("it","../dataset/cleaned_tweets/cleaned_italian_tweets.csv")
all_events = evaluator.get_all_events("../dataset/event_extracted/nltk/event_extracted_italian_nltk.csv")
accuracy_nltk = evaluator.accuracy(all_events)
print(f'Accuracy: {accuracy_nltk*100}%')
```

```python
number_of_events_extracted_nltk = evaluator.get_n_events_classified()
```

```python
labels = ["Stanza", "Spacy", "Nltk"]
lengths = [accuracy_stanza, accuracy_spacy, accuracy_nltk]

plt.figure(figsize=(10, 6))
bars = plt.bar(labels, lengths, width=0.6)

title = "Comparison between NER packages on Italian datasets"

plt.title(title, fontsize=16)
plt.ylabel('Accuracy', fontsize=12)
plt.ylim(0, max(lengths) * 1.1) 


    
bars[0].set_color('skyblue')
bars[1].set_color('lightgreen')

fig1 = plt.gcf()
plt.show()
plt.draw()
fig1.savefig(title, dpi=100)
```

## Comparison on English datasets

```python
evaluator = perfomance_evaluator.PerformanceEvaluator("en","../dataset/cleaned_tweets/cleaned_english_tweets.csv")
all_events = evaluator.get_all_events("../dataset/event_extracted/stanza/event_extracted_english_stanza.csv")
accuracy_stanza_en = evaluator.accuracy(all_events)
print(f'Accuracy: {accuracy_stanza_en*100}%')
```

```python
evaluator = perfomance_evaluator.PerformanceEvaluator("en","../dataset/cleaned_tweets/cleaned_english_tweets.csv")
all_events = evaluator.get_all_events("../dataset/event_extracted/spacy/event_extracted_english_spacy.csv")
accuracy_spacy_en = evaluator.accuracy(all_events)
print(f'Accuracy: {accuracy_spacy_en*100}%')
```

```python
evaluator = perfomance_evaluator.PerformanceEvaluator("en","../dataset/cleaned_tweets/cleaned_english_tweets.csv")
all_events = evaluator.get_all_events("../dataset/event_extracted/nltk/event_extracted_english_nltk.csv")
accuracy_nltk_en = evaluator.accuracy(all_events)
print(f'Accuracy: {accuracy_nltk_en*100}%')
```

```python
labels = ["Stanza", "Spacy", "Nltk"]
lengths = [accuracy_stanza_en, accuracy_spacy_en, accuracy_nltk_en]

plt.figure(figsize=(10, 6))
bars = plt.bar(labels, lengths, width=0.6)

title = 'Comparison between NER packages on English datasets'
plt.title(title, fontsize=16)
plt.ylabel('Accuracy', fontsize=12)
plt.ylim(0, max(lengths) * 1.1) 


bars[0].set_color('skyblue')
bars[1].set_color('lightgreen')

fig1 = plt.gcf()
plt.show()
plt.draw()
fig1.savefig(title, dpi=100)
```

```python
X = ['Stanza','Spacy','Nltk'] 
accuracy_en = [accuracy_stanza_en,accuracy_spacy_en,accuracy_nltk_en] 
accuracy_it = [accuracy_stanza,accuracy_spacy,accuracy_nltk] 
  
X_axis = np.arange(len(X)) 
  
plt.bar(X_axis - 0.2, accuracy_en, 0.4, label = 'English') 
plt.bar(X_axis + 0.2, accuracy_it, 0.4, label = 'Italian') 

title = "Comparison between languages"
plt.xticks(X_axis, X) 
plt.xlabel("NER tools") 
plt.ylabel("Accuracy") 
plt.title(title) 
plt.legend() 

fig1 = plt.gcf()
plt.show()
plt.draw()
fig1.savefig(title, dpi=100)
```

```python
X = ['Stanza','Spacy','Nltk'] 
events_groundtruth = [number_of_events_groundtruth,number_of_events_groundtruth,number_of_events_groundtruth] 
events_extracted = [number_of_events_extracted_stanza,number_of_events_extracted_spacy,number_of_events_extracted_nltk] 
  
X_axis = np.arange(len(X)) 
  
plt.bar(X_axis - 0.2, events_groundtruth, 0.4, label = 'Groundtruth') 
plt.bar(X_axis + 0.2, events_extracted, 0.4, label = 'Event Extracted') 

title = "Number of event extracted with different NER packages"
plt.xticks(X_axis, X) 
plt.xlabel("NER tools") 
plt.ylabel("Number") 
plt.title(title) 
plt.legend() 

fig1 = plt.gcf()
plt.show()
plt.draw()
fig1.savefig(title, dpi=100)
```
