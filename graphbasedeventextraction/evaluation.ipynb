```python
%matplotlib inline
import matplotlib.pyplot as plt
from util import perfomance_evaluator 
```

```python
evaluator = perfomance_evaluator.PerformanceEvaluator("it","../dataset/cleaned_tweets/cleaned_italian_tweets.csv")
```

```python
all_events = evaluator.get_all_events("../dataset/event_extracted/stanza/event_extracted_italian_stanza.csv")
```

```python
accuracy = evaluator.accuracy(all_events)
print(f'Accuracy: {accuracy*100}%')
```

```python
number_of_events_groundtruth = evaluator.get_n_events_groundtruth()
number_of_events_extracted = evaluator.get_n_events_classified()

labels = ["Real Event", "Event Extracted"]
lengths = [number_of_events_groundtruth, number_of_events_extracted]

plt.figure(figsize=(10, 6))
bars = plt.bar(labels, lengths, width=0.6)

title = 'Comparison between Groundtruth and Clustering'
plt.title(title, fontsize=16)
plt.ylabel('Number of Events', fontsize=12)
plt.ylim(0, max(lengths) * 1.1) 

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height}',
             ha='center', va='bottom')
    
bars[0].set_color('skyblue')
bars[1].set_color('lightgreen')

fig1 = plt.gcf()
plt.show()
plt.draw()
fig1.savefig(title, dpi=100)
```

```python
n_what, n_who, n_where, n_when = evaluator.check_named_entity(all_events)
labels = ["What", "Who","Where","When"]
lengths = [n_what, n_who, n_where,n_when]

plt.figure(figsize=(10, 6))
bars = plt.bar(labels, lengths, width=0.6)

title = 'Tendency of Named Entities to be non-empty'
plt.title(title, fontsize=16)
plt.ylabel('Percentage', fontsize=12)
plt.ylim(0, max(lengths) * 1.1) 


bars[0].set_color('blue')
bars[1].set_color('orange')
bars[2].set_color('red')
bars[3].set_color('green')

fig1 = plt.gcf()
plt.show()
plt.draw()
fig1.savefig(title, dpi=100)
```
